```python
#Importing libraries that we need
from faker import Faker
import pandas as pd
import datetime
import random

#create an instance for faker
fake=Faker(locale='en_US')

#Function to create dataset
def personal_expenses(num):
    S_num=[] #created empty list to add our details
#creating an expense dictionary
    for i in range(1,num):
        expense={}
        start_date = datetime.date(2024, 1, 1)
        end_date = datetime.date(2024, 1, 31)
        expense['Date'] = fake.date_between(start_date=start_date, end_date=end_date)
        expense['Category']= fake.random_element(elements=("Investment","School fees","Groceries","Loan amount","Shopping","Subscription","Health & Medicine"))
        expense['Payment_Mode']= fake.random_element(elements=("UPI","Cash","Netbanking","Credit card","Debit card","Wallet"))
        def discrip(Category):
          if Category == "Investment":
            return random.choice(["Investment on stock","Investment on Bank"])
          elif Category =="School fees":
            return "Fees for School"
          elif Category=="Groceries":
            return random.choice(["Home essential purchase","Fruits & Vegetables purchase"])
          elif Category=="Loan amount":
            return "Housing loan paid"
          elif Category =="Shopping":
            return random.choice(["Dress purchase","Cosmetic Purchase","Movie charge","Toys/Gifts Purchase"])
          elif Category=="Subscription":
            return random.choice(["Subscription for Hotstar","Subscription for Netflix"])
          elif Category=="Health & Medicine":
            return random.choice(["Regular Medicine Purchase","Hospitalization"])
          else:
            return fake.sentence()
        expense['Discription']= discrip(expense['Category'])
        expense['Amount']= round(fake.random.uniform(100,5000),2)
        expense['CashBack']= round(fake.random.uniform(0,7),2)
        S_num.append(expense)
    return pd.DataFrame(S_num)

personal_expenses(15)

#det=personal_expenses(32)
#det.to_csv("personal_expenses.csv")
```
